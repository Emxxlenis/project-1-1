        Función jugar_sudoku(tablero_original, tablero_juego)
            Imprimir mensaje de bienvenida

            Mientras Verdadero hacer
                Imprimir tablero_juego usando imprimir_matriz

                Intentar
                    Solicitar fila al usuario y convertir a entero
                    Solicitar columna al usuario y convertir a entero

                    Si fila o columna están fuera de rango (0-3) entonces
                        Imprimir mensaje de error

                    FinSi

                        Imprimir mensaje de error
                    FinSi

                    Si tablero_completo(tablero_juego) entonces
                        Imprimir mensaje de victoria

                        Romper ciclo

                Capturar error de entrada inválida

```python
import random
# Definimos 5 tableros de Sudoku de 4x4 en nivel fácil (previamente resueltos)
tableros = [
    [
        [1, 4, 3, 2],
        [3, 2, 1, 4],
        [4, 3, 2, 1],
        [2, 1, 4, 3]
    ],
    [
        [2, 3, 4, 1],
        [4, 1, 3, 2],
        [1, 4, 2, 3],
        [3, 2, 1, 4]
    ],
    [
        [3, 2, 4, 1],
        [1, 4, 2, 3],
        [2, 1, 3, 4],
        [4, 3, 1, 2]
    ],
    [
        [4, 1, 2, 3],
        [3, 2, 4, 1],
        [1, 3, 4, 2],
        [2, 4, 1, 3]
    ],
    [
        [1, 3, 4, 2],
        [2, 4, 1, 3],
        [3, 1, 2, 4],
        [4, 2, 3, 1]
    ]
]

def modificar_tablero(tablero):
    """
    Crea una versión modificada del tablero original, ocultando el 50% de las celdas con '_'.

    """
    tablero_modificado = [fila[:] for fila in tablero]  # Copia profunda del tablero para evitar modificar el original
    filas = len(tablero_modificado)
    columnas = len(tablero_modificado[0])
    total_celdas = filas * columnas
    celdas_a_reemplazar = total_celdas // 2  # Se oculta aproximadamente el 50% de las celdas

    # Ocultamos celdas aleatorias hasta alcanzar el 50% del tablero
    celdas_ocultas = set()
    while len(celdas_ocultas) < celdas_a_reemplazar:
        i, j = random.randint(0, filas - 1), random.randint(0, columnas - 1)
        if (i, j) not in celdas_ocultas:
            tablero_modificado[i][j] = "_"
            celdas_ocultas.add((i, j))  # Evitamos reemplazar la misma celda varias veces

    return tablero_modificado

def imprimir_matriz(matriz):
    """
    Imprime una matriz de manera legible en la consola.

    """
    for fila in matriz:
        print(" ".join(map(str, fila)))  # Convierte los valores a string y los une con espacios

def tablero_completo(tablero):
    """
    Verifica si el tablero está completamente lleno, es decir, si no hay espacios vacíos ('_').

    """
    return all("_" not in fila for fila in tablero)  # Retorna True si no hay '_' en ninguna fila

def jugar_sudoku(tablero_original, tablero_juego):
    """
    Inicia el juego de Sudoku permitiendo al usuario ingresar valores en las celdas vacías.

    """
    print("\n¡Bienvenido a Batalla Sudoku!")
    print("Resuelve el tablero llenando los espacios vacíos.")

    while True:
        print("\nTablero Actual:")
        imprimir_matriz(tablero_juego)

        try:
            # Solicitar la posición al usuario
            fila = int(input("Ingrese la fila (horizontal) (0-3): "))
            col = int(input("Ingrese la columna (vertical) (0-3): "))

            # Validar que la posición esté dentro del rango permitido
            if not (0 <= fila < 4 and 0 <= col < 4):
                print("⚠️ Posición fuera de rango. Intente de nuevo.")
                continue

            # Verificar si la celda ya tiene un número
            if tablero_juego[fila][col] != "_":
                print("⚠️ Esta posición ya tiene un número. Elija otra.")
                continue

            # Solicitar el número al usuario
            num = int(input("Ingrese un número (1-4): "))

            # Validar si el número ingresado es correcto
            if num == tablero_original[fila][col]:
                tablero_juego[fila][col] = num  # Se actualiza la celda si es correcto
                print("✅ ¡Correcto!")
            else:
                print("❌ Incorrecto. Intenta otra vez.")

            # Comprobar si el usuario ha completado el tablero
            if tablero_completo(tablero_juego):
                print("\n🎉 ¡Felicidades! Has completado el Sudoku.")
                imprimir_matriz(tablero_juego)
                break

        except ValueError:
            print("⚠️ Entrada inválida. Ingrese solo números.")

# Seleccionar un tablero al azar de la lista de tableros
tablero_original = random.choice(tableros)
tablero_juego = modificar_tablero(tablero_original)

# Iniciar el juego de Sudoku
jugar_sudoku(tablero_original, tablero_juego)


```
